## From an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data, on_bad_lines='skip')
```

#### > Rename function

```python
def rename(city):
    if city == 'city_103':
        return "Jaipur"
    else:
        return city
```

```python
rename('city_103')
```

```python
pd.read_csv('/kaggle/input/aug-train/aug_train.csv', converters={'city': rename})
```

## Encoding Parameter

#### > Avoid UnicodeDecodeError

```python
pd.read_csv('/kaggle/input/zomato-dataset/zomato.csv', encoding = 'latin-1')
```

> If you're unsure about the encoding, you may need to try multiple encodings until you find the one that works without errors or strange characters in your data.

```python
# List of encodings to try
encodings_to_try = ['utf-8', 'latin-1', 'ISO-8859-1', 'cp1252']

# Try different encodings until one works without errors
for encoding in encodings_to_try:
    try:
        df = pd.read_csv('/kaggle/input/zomato-dataset/zomato.csv', encoding=encoding)
        print("File successfully read with encoding:", encoding)
        break
    except UnicodeDecodeError:
        print("Failed to read with encoding:", encoding)
```
